# Text classification Project

##                              Necessary modules to import

```python
import numpy as np

import nltk
from nltk.tokenize import word_tokenize  #to split words from string
from nltk.corpus import stopwords

from string import punctuation

from sklearn.model_selection import train_test_split

import os #used for reading file from directory
from nltk.stem import WordNetLemmatizer
from collections import Counter

from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import   MultinomialNB
```


 ##                              Features selecting work 


```python
lemmatizer=WordNetLemmatizer()
```

```python
#adding stopword in stops


stops=stopwords.words('english')
stops=stops+list(punctuation)
stops.extend(['','--'])
stops
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else :
        return  wordnet.NOUN
```


# COLLECTING ALL WORDS  FROM ALL DOCUMENT IN  #all_words[ ] after cleaning them  and then selecting top 2500 as the features


```python
all_words=[]
count=1
dirs=os.listdir('20_newsgroups')

for directory in dirs:                     #each folder of 20_newsgroups
    path='20_newsgroups/'+directory        #creating path of each foldere
    files=os.listdir(path)                 #inside each folder
    
    print(count ,' ', ' hello')            #it will print hello when each direcotry start to indicate simply that our code is 
    count=count+1                          #running bcz
    
    for file in files[0:800]:              #picking each cateogries only first 800 files all other are used for testing
        path_of_file_in_dir=path+'/'+file  #path of each file
        ptr=open(path_of_file_in_dir)      #opening file    
        text=ptr.read()
        words=word_tokenize(text)       #splitting text into words
        
        for each_word in words:
            clean_words=lemmatizer.lemmatize( each_word,pos= get_simple_pos(each_word ))
            if clean_words.lower() not in stops:
                all_words.append(clean_words.lower())
```

```python
result=nltk.FreqDist(all_words)
```

```python
features = result.most_common(3000)
features
```


# now i have  2000 features


```python
final_features=[i[0] for i in features]
```


## making xtrain and ytrain and dictionary of list of counters


```python

ytrain=[]  # type of doucument   
xtrain=[]  #i will append another array that will define the frequency of each FINAL_FEATURES in document  
#will store the frequency of first 800 files of each directory


filecount=[]   # for each document i  will store the counter in it i.e list of counters

count=1
dirs=os.listdir('20_newsgroups')

for directory in dirs:
    path='20_newsgroups/'+directory
    files=os.listdir(path)
    print(count,'hello')
    count=count+1
    for file in files[0:750]:
        ytrain.append(directory)
        newpath=path+'/'+file
        ptr=open(newpath)           #opening file   
        text=ptr.read()
        text=word_tokenize(text)
        c=Counter()
        for word in text:
            c[word]=c[word]+1
        filecount.append(c)
        
#NOW TO STORE THE FREQUNCY OF EACH FETURES FOR EACH FILE         
for eachcounter in filecount:
    subxtrain=[]
    for eachfeature in final_features:
        subxtrain.append( eachcounter[eachfeature] )
    xtrain.append(subxtrain)
```

```python
#Now to build a dictionary that have dict['class']->['each feature']=its frequency count

def fit():
    dictionary={}
    classes=set(ytrain)
    dictionary['total_length']=len(xtrain)
    
    for curr_class in classes :
        print( 'working on ',curr_class)
        dictionary[curr_class] = {}
        curr_class_rows = (ytrain==curr_class)
        curr_class_y = ytrain[curr_class_rows]
        curr_class_x = xtrain[curr_class_rows]
        dictionary[curr_class]['total_length']=len(curr_class_y)
        
        for i  in range(1,len(final_features)+1):
            dictionary[curr_class]['total_words']= curr_class_x.sum()    #total words in currentClass X
            dictionary[curr_class][i]= curr_class_x[:,i-1].sum( )         #total words ith feature in class
            
    return dictionary
```

```python
def probability(curr_class,x,dictionary):
    
    # p(y=curr_class/ x=x) = p(x=x/y=curr_class ) * p(y=curr_class)
    # p(y=curr_class)=  ( no of data that has class y / total data  )
    
    num= dictionary[curr_class]['total_length']
    den= dictionary['total_length']
    result=np.log(num)-np.log(den)
    
    for i in range(1,len(final_features)+1) :
        #p(x=x/y=curr_class)=(no of data has class y for featues x +1 )/( total data in curr_class + len(final_features)  ) 
        
        if(x[i-1]!=0):
            
            num2=1+dictionary[curr_class][i]
            
            den2=len(final_features) +dictionary[curr_class]['total_words']
            
                result=result+x[i-1]*( np.log(num2)-np.log(den2) )
            
    return result 
```

```python
def predict_for_single(dictionary,x):
    
    #result_class='hello'  #i will store in this result class 
    
    result_prob=-100000
    classes=set(ytrain)
    flag=0  #for first iteration
    
    for curr_class in classes:
        if(curr_class=='total_length'):
            continue
        temp=probability(curr_class,x,dictionary)
        if(flag==0 or temp>result_prob):
            flag=1
            result_prob=temp
            result_class=curr_class    
            
    return result_class
```

```python
def predict(dictionary,ytest):
    ypred=[]
    count2=1 
    for x in xtest:
        print('finding result for ', count2)           # Just to indicate that program is running
        count2=count2+1                                
        ypred.append(predict_for_single(dictionary,x))
    return ypred
```

```python
#ypred contain the result classes for ytest

dictionary1=fit()
```
